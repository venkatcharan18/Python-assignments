```python
import random
def roll():
    return random.randint(1,6)

p1=0
p2=0
while True:
    
#PLAYER 1-
    input('press a key to roll dice: ')
    dice_1=roll()
    print('player 1 rolled',dice_1)
    
    if p1==0:
        if dice_1==6:
            p1=1        
            print('player 1 moved to pos 1')
    else:
        p1+=dice_1
        if   p1 == 8:  # ladder
            p1 = 36
            print('player 1 position is ',p1)
        elif p1 == 55:  # ladder
            p1 = 80
            print('player 1 position is ',p1)
        elif p1 == 35:  # snake
            p1 = 26
            print('player 1 position is ',p1)
        elif p1 == 67:  # snake
            p1 = 40
            print('player 1 position is ',p1)
        elif p1>100:
            p1=p1-dice_1
            print('player 1 position is ',p1)
        elif p1==100:
            print('player 1 win')
            break
        else:
            print('player 1 position is ',p1)
            
#PLAYER 2 -           

    input ('press a key to roll dice: ')
    dice_2=roll()
    print('player 2 rolled',dice_2)
    
    if p2==0:
        if dice_2==6:
            p2=1
            print('player 2 moved to pos 1')
    else:
        p2+=dice_2
        if   p2 == 8:  # ladder
            p2 = 36
            print('player 2 position is ',p2)
        elif p2 == 55:  # ladder
            p2 = 80
            print('player 2 position is ',p2)
        elif p2 == 35:  # snake
            p2 = 26
            print('player 2 position is ',p2)
        elif p2 == 67:  # snake
            p2 = 40
            print('player 2 position is ',p2)
        elif p2>100:
            p2=p2-dice_2
            print('player 2 position is ',p2)
        elif p2==100:
            print('player 2 win')
            break
        else:
            print('player 2 position is ',p2)
    
    

    
    
        
```
